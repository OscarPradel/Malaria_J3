To better understand the difference between the .png images of the infected and the healthy cells, we will plot some samples below:

```python
plt.figure(figsize = (15,15))
# Loop to plot 5 images of uninfected cells
for i in range(5):
    plt.subplot(1, 5, i+1)
    img = cv2.imread('/home/itmalgrp04/cell_images/Uninfected' + "/" + uninfected_data[i+1])
    plt.title('Uninfected')
    plt.imshow(img)
plt.show()

plt.figure(figsize = (15,15))
# Loop to plot 5 images of infected cells
for i in range(5):
    plt.subplot(1, 5, i+1)
    img = cv2.imread('/home/itmalgrp04/cell_images/Parasitized' + "/" + parasitized_data[i])
    plt.title('Parasitized')
    plt.imshow(img)
plt.show()
```

Looking at the sample plots above, it is clear to see, that the images dont scale the same way. this could lead to issues. To fix this, we proceed to resizing the images the exact same way:

```python
data = []
labels = []
for img in parasitized_data:
    try:
        # Reading image
        img_read = plt.imread('/home/itmalgrp04/cell_images/Parasitized' + "/" + img)
        # Resizing image
        img_resize = cv2.resize(img_read, (50, 50))
        # Convert a PIL image to a Numpy array
        img_array = img_to_array(img_resize)
        # Adding to end of list
        data.append(img_array)
        # Adding to labels list. Value 1 when Parasitized, 0 when Uninfected
        labels.append(1)
    except:
        None
        
for img in uninfected_data:
    try:
        img_read = plt.imread('/home/itmalgrp04/cell_images/Uninfected' + "/" + img)
        img_resize = cv2.resize(img_read, (50, 50))
        img_array = img_to_array(img_resize)
        data.append(img_array)
        labels.append(0)
    except:
        None
```

```python
image_data = np.array(data)
labels = np.array(labels)
```

```python
idx = np.arange(image_data.shape[0])
np.random.shuffle(idx)
image_data = image_data[idx]
labels = labels[idx]
```

```python
# Splitting into test data:
# FIND EXPLAINATION AND REFACTOR

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(image_data, labels, test_size = 0.2, random_state = 42)

```

```python
y_train = np_utils.to_categorical(y_train, num_classes = 2)
y_test = np_utils.to_categorical(y_test, num_classes = 2)
```

```python
# REFACTOR AND EXPLAIN
print(f'SHAPE OF TRAINING IMAGE DATA : {x_train.shape}')
print(f'SHAPE OF TESTING IMAGE DATA : {x_test.shape}')
print(f'SHAPE OF TRAINING LABELS : {y_train.shape}')
print(f'SHAPE OF TESTING LABELS : {y_test.shape}')
```

```python
#BUILDING CNN MODEL without data augmentation
#REFACTOR THIS

def CNNbuild(height, width, classes, channels):
    model = Sequential()
    
    inputShape = (height, width, channels)
    
    if K.image_data_format() == 'channels_first':
        inputShape = (channels, height, width)
    model.add(Conv2D(32, (3,3), activation = 'relu', input_shape = inputShape))
    model.add(MaxPooling2D(2,2))

    model.add(Conv2D(32, (3,3), activation = 'relu'))
    model.add(MaxPooling2D(2,2))

    model.add(Flatten())
    
    model.add(Dense(128, activation = 'relu'))
    model.add(Dense(classes, activation = 'softmax'))
    
    return model
```

```python
# Instantiate model
# Refactor this

height = 50
width = 50
classes = 2
channels = 3
model = CNNbuild(height = height, width = width, classes = classes, channels = channels)
model.summary()
```

```python
#compile the model
model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

```python
#fit the model onto the dataset
modelFit = model.fit(x_train, y_train, epochs = 20, batch_size = 32)
```

```python
plt.figure(figsize = (15,5))
plt.title('Training Accuracy and Loss')
plt.ylabel('Accuracy and Loss')
plt.xlabel("Epochs")

plt.plot(range(20), modelFit.history['loss'], label = 'Taining Loss')
plt.plot(range(20), modelFit.history['acc'], label = 'Training Accuracy')
```

```python
#evaluate the model on test data
evaluation = model.evaluate(x_test, y_test)
print(f'Accuracy: {evaluation[1]} Loss: {evaluation[0]}')
```

27558 images isnt alot of data to train a model, so data augmentation might help

```python
#Refactor
# https://www.kaggle.com/gimunu/data-augmentation-with-keras-into-cnn
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1/255.,
                                  horizontal_flip = True,
                                  width_shift_range = .15,
                                  height_shift_range = .15,
                                  fill_mode = 'nearest',
                                  zoom_range = 0.1,
                                  rotation_range = 15)
val_datagen = ImageDataGenerator(rescale = 1/255.)

train_generator = train_datagen.flow(x_train, y_train, batch_size = 64, shuffle = False)
val_generator = val_datagen.flow(x_test, y_test, batch_size = 64, shuffle = False)
```

```python
#refactor

#calling the same model as above
model_aug = CNNbuild(height = height, width = width, classes = classes, channels = channels)
```

```python
#Refactor

#compile the model
optim = optimizers.Adam(lr = 0.001, decay = 0.001 / 64)
model_aug.compile(loss = 'categorical_crossentropy', optimizer = optim, metrics = ['accuracy'])
```

```python
#Refactor
#fit the model on the augmented dataset
h_aug = model_aug.fit_generator(train_generator, steps_per_epoch = len(x_train) // 64, epochs = 50)
```

```python
#refactor
#evaluate the model on augmented test data
predict = model_aug.evaluate_generator(val_generator, steps = 5)
```

```python
#Refactor
print(f'LOSS ON TEST DATA AFTER DATA AUGMENTATION : {predict[0]}')
print(f'ACCURACY ON TEST DATA AFTER DATA AUGMENTATION : {predict[1]}')
```
